```python
from PIL import Image

def clipping(intensitas):
  if intensitas < 0:
    return 0
  if intensitas > 255:
    return 255
  return intensitas

def atur_pencerahan(nilai_pencerahan, nama_setelah_disave):
  CITRA = Image.open('pictogram-884043_1280.PNG')
  PIXEL = CITRA.load()

  ukuran_horizontal = CITRA.size[0]
  ukuran_vertikal = CITRA.size[1]

  # The for loops should be indented to be part of the function
  for x in range(ukuran_horizontal):
    for y in range(ukuran_vertikal):
      R = clipping(PIXEL[x, y][0] + nilai_pencerahan)
      G = clipping(PIXEL[x, y][1] + nilai_pencerahan)
      B = clipping(PIXEL[x, y][2] + nilai_pencerahan)
      PIXEL[x, y] = (R, G, B)

  CITRA.save(nama_setelah_disave)

atur_pencerahan(128, 'pictogram-884043_1280_dicerahkan.PNG')
atur_pencerahan(-128, 'pictogram-884043_1280_digelapkan.PNG')
```

```python
# Filter linier
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
image = cv.imread("disabilitas.png")
fig, ax = plt.subplots(1, 3, figsize=(16, 8))
fig.tight_layout()
# untuk meng-konvolusi kernel pada sebuah gambar kita dapat menggunakan cv.filter2D
kernel_sharpening = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
kernel_sharpening_2 = np.array([[-1, -1, -1], [-1, 10, -1], [-1, -1, -1]])

ax[0].imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
sharpened = cv.filter2D(image, -1, kernel_sharpening)
ax[1].imshow(cv.cvtColor(sharpened, cv.COLOR_BGR2RGB))
ax[1].set_title('Sharpened Kernel Image')
sharpened_2 = cv.filter2D(image, -1, kernel_sharpening_2)
ax[2].imshow(cv.cvtColor(sharpened_2, cv.COLOR_BGR2RGB))
ax[2].set_title('Sharpened Kernel Image 2')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
image = cv.imread("disabilitas.png")
h, w = image.shape[:2]
half_height, half_width = h//4, w//8
transition_matrix = np.float32([[1, 0, half_width], [0, 1, half_height]])
img_transition = cv.warpAffine(image, transition_matrix, (w, h))
plt.imshow(cv.cvtColor(img_transition, cv.COLOR_BGR2RGB))
plt.title("Translation")
plt.show()
```

```python
import cv2
import numpy
img = cv2.imread ('disabilitas.png' , cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5,5),np.uint8)
from google.colab.patches import cv2_imshow
cv2_imshow(img)
#first erosi
erosi = cv2.erode(img, kernel, iterations=1)
from google.colab.patches import cv2_imshow
cv2_imshow(erosi)
#second dilasi
dilasi = cv2.dilate (img,kernel,iterations=1)
from google.colab.patches import cv2_imshow
cv2_imshow(dilasi)
#thrid Opening yakni = erosi dahulu lalu dilasi
opening = cv2.morphologyEx (img, cv2.MORPH_OPEN, kernel)
from google.colab.patches import cv2_imshow
cv2_imshow(opening)

#fourth Closing yakni = dilasi dahulu lalu erosi
closing = cv2.morphologyEx (img, cv2.MORPH_CLOSE, kernel)
from google.colab.patches import cv2_imshow
cv2_imshow(closing)
#ini Gradien beda antara dilasi dan erosi
gradient = cv2.morphologyEx (img, cv2.MORPH_GRADIENT, kernel)
from google.colab.patches import cv2_imshow
cv2_imshow(gradient)
#ini Topi Atas beda antara opening dan closing
tophat = cv2.morphologyEx (img, cv2.MORPH_TOPHAT, kernel)
from google.colab.patches import cv2_imshow
cv2_imshow(tophat)
#ini Topi black beda antara closing dan opening
blackhat = cv2.morphologyEx (img, cv2.MORPH_BLACKHAT, kernel)
from google.colab.patches import cv2_imshow
cv2_imshow(blackhat)
#https://docs.opencv.org/4.x/d9/d61/tutorial_py_morphological_ops.html
```
